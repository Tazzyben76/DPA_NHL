```python
# Declare parameters 
project_name = "nhl" # Name of the project (used for folder structure)
kaggle_dataset = "martinellis/nhl-game-data" # Kaggle dataset identifier (owner/dataset-name)
```

```python
# Configure basic logging for visibility in Fabric runs
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("extract")
```

```python
# Initialize Kaggle API client
api = KaggleApi()
api.authenticate()

logger.info("Authenticated with Kaggle API")
```

```python
# Absolute filesystem path inside the Fabric Lakehouse
raw_fs_path = f"/lakehouse/default/Files/Raw/{project_name}"

# Create folder if it does not exist
os.makedirs(raw_fs_path, exist_ok=True)

logger.info(f"Raw data folder ready: {raw_fs_path}")
```

```python
try:
    # Download and unzip all files from the Kaggle dataset
    api.dataset_download_files(
        kaggle_dataset,
        path=raw_fs_path,
        unzip=True
    )
    logger.info("Dataset successfully downloaded and extracted")

except Exception:
    # Log full stack trace if anything fails
    logger.exception("Kaggle dataset download failed")
    raise
```
