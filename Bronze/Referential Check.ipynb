```python
from pyspark.sql import functions as F

# Function to do referential check, ensure foreign key validity in the corresonding reference table's primary key

def check_fk(
    child_df,
    ref_df,
    child_col,
    ref_col,
    child_table_name,
    ref_table_name
):
    violations = (
        child_df
        .select(child_col)
        .distinct()
        .join(
            ref_df.select(ref_col).distinct(),
            child_df[child_col] == ref_df[ref_col],
            how="left_anti"
        )
    )

    print(f"\nğŸ” FK CHECK: {child_table_name}.{child_col} â†’ {ref_table_name}.{ref_col}")
    print(f"âŒ Violations count: {violations.count()}")

    violations.show(100, truncate=False)

    return violations
```

```python
# Load reference tables (Bronze)

nhl_game = spark.read.table("NHL_Lakehouse_Bronze.nhl_game")
nhl_team_info = spark.read.table("NHL_Lakehouse_Bronze.nhl_team_info")
nhl_player_info = spark.read.table("NHL_Lakehouse_Bronze.nhl_player_info")
nhl_game_plays = spark.read.table("NHL_Lakehouse_Bronze.nhl_game_plays")
```
